```python
import requests
from bs4 import BeautifulSoup
import re
import sqlite3
import pandas as pandas
fdic = {}
list_a = []
list_b = []
url = 'https://fine.fss.or.kr/fine/fnctip/fncDicary/list.do?menuNo=900021&pageIndex='
print(url)
```

```python
response = requests.get(url)
```

```python
list_a.clear()
list_b.clear()
```

```python
if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    ul = soup.select_one('div.result-list')
    print(ul)
    sub=ul.select(' .bd-list.result-list>  dl>   dt')
    con = ul.select('  .bd-list.result-list> dl>  dd')
    for i in sub:
        tmp = i.get_text().strip()
        tmp = tmp.replace("\r\n\t\t\t\t\t\t\t\t\xa0", "")
        tmp = tmp.replace("\r\n\t\t\t\t","")
        tmp = tmp.replace("\t","")
        tmp = re.sub('^[0-9]+. ', '', tmp)
        list_a.append(tmp)
    for i in con:
        tmp = i.get_text().strip()
        tmp = tmp.replace("\r\n\t\t\t\t","")
        tmp = re.sub('^[0-9]+. ', '', tmp)
        list_b.append(tmp)
else:
    print(response.status_code)
    
```

```python
import pandas as pd
df = pd.DataFrame(zip(list_b, list_a))
```

```python
conn=sqlite3.connect('d:/Programfiles/sqlite3/fdic.db')
```

```python
cur=conn.cursor()
```

```python
conn.execute('create table fss_dic(id integer, name text, content text)')
```

```python
for i in fdic:
    name = fdic[i][0]
    content = fdic[i][1]
    sql = 'insert into fss_dic values (?, ?, ?)'
    cur.execute(sql, (i, name, content))
```
